# Visualization of junco bird pattern

```python
import numpy as np
import pandas as pd
import folium
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("junco_flickr.csv", delimiter = "|")
# read the input dataset
```

```python
df.head()
```

```python
loc_df = df.query("Latitude != 0 & Longitude != 0")
```

```python
junco_map = folium.Map(tiles = "Stamen Terrain")
```

```python
def add_to_map(row):
    
    loc = (row["Latitude"], row["Longitude"])
    folium.Marker(location = loc, ).add_to(junco_map)
```

```python
loc_df.apply(add_to_map, axis = 1)
junco_map
```
